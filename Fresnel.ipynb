```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,33*np.pi/64,np.pi/64)#angulo de incidencia
n = 1.51
```

```python
r_tm = ( -(n**2) * np.cos(t) + np.sqrt((n**2)-(np.sin(t))**2) ) / ( (n**2)*np.cos(t) + np.sqrt((n**2)-(np.sin(t))**2) )
r_te = ( np.cos(t) - np.sqrt((n**2)-(np.sin(t))**2) ) / ( np.cos(t) + np.sqrt((n**2)-(np.sin(t))**2) )
t_tm = ( 2*np.cos(t) ) / ( (n**2)*np.cos(t) + np.sqrt((n**2)- (np.sin(t))**2) )
t_te = ( 2*np.cos(t) ) / ( np.cos(t) + np.sqrt((n**2)- (np.sin(t))**2) )
```

```python
plt.figure(figsize = (8,8))
plt.title(r'Coeficientes de Fresnel $\Gamma_{TE}$, $\Gamma_{TM}$, $T_{TE}$, $T_{TM}$')
plt.plot(t,r_te,label = r'$\Gamma_{TE}$')
plt.plot(t,r_tm,label = r'$\Gamma_{TM}$')
plt.plot(t,t_te,label = r'$T_{TE}$')
plt.plot(t,t_tm,c = 'black',label = r'$T_{TM}$')
plt.axvline(np.pi/2,ls = ':')
plt.plot(np.pi*56.31/180,0,'bo',label='Brewsters Angle')
plt.ylabel('Valor del Coeficiente',fontsize = 15)
plt.xlabel(r'Angulo de Incidencia $\theta_i$',fontsize = 15)
plt.legend(fontsize = 17)
plt.grid()
```
